```python
for item in data:
    if data[item].dtype in ['float64', 'int64']:
        print (item)
        print ("Среднее: ", statistics.mean(data[item]))
        print ("Медиана: ", statistics.median(data[item]))
        #print ("Мода: ", statistics.mode(data[item]))
        print ("Отклонение: ", statistics.pstdev(data[item]))
        print ("Минимум: ", min(data[item]))
        print ("Максимум: ", max(data[item]))
        print ("-----------")
        
```

```python
def getIntSalary (row):
    if row['salary'] == 'low':
        return 0
    if row['salary'] == 'medium':
        return 1
    return 2

data['salaryInt'] = data.apply(lambda row: getIntSalary(row), axis=1)
```

```python
from sklearn import preprocessing

dataIntIndex = data.columns[data.dtypes.apply(lambda c: np.issubdtype(c, np.number))]
dataInt = data[dataIntIndex]

min_max_scaler = preprocessing.MinMaxScaler(feature_range=(-1,1))
dataNorm = pd.DataFrame(min_max_scaler.fit_transform(dataInt))
dataNorm.columns = dataIntIndex
dataNorm.head()
```

```python
import seaborn as sns
sns.heatmap(dataNorm.corr(),annot=True,cmap='RdYlGn',linewidths=0.2)
fig=plt.gcf()
fig.set_size_inches(12,12)
plt.show()
```

```python
#Нулевая гипотеза о том, что у уволенных и работающих нет разницы в количестве проектов, проводимых за время работы
data['projectsByTime'] = data['average_montly_hours']/data['number_project']
fired = data[data['left']==1]['projectsByTime']
working = data[data['left']==0]['projectsByTime']
stats.ttest_ind(fired, working, equal_var=False)

#Нулевая гипотеза отвергается, значит, одна из причин увольнения - трата большого кол-ва времени на малое кол-во проектов, т.е. нет разнообразия или развития
```
