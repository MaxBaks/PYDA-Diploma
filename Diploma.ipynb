3.Рассчитать и визуализировать корреляционную матрицу для переменных.

```python
def getIntSalary (row):
    if row['salary'] == 'low':
        return 1
    if row['salary'] == 'medium':
        return 2
    return 3

data['salaryInt'] = data.apply(lambda row: getIntSalary(row), axis=1)
```

```python
from sklearn import preprocessing

dataIntIndex = data.columns[data.dtypes.apply(lambda c: np.issubdtype(c, np.number))]
dataInt = data[dataIntIndex]

min_max_scaler = preprocessing.MinMaxScaler(feature_range=(-1,1))
dataNorm = pd.DataFrame(min_max_scaler.fit_transform(dataInt))
dataNorm.columns = dataIntIndex
dataNorm.head()
```

```python
import seaborn as sns
sns.heatmap(dataNorm.corr(),annot=True,cmap='RdYlGn',linewidths=0.2)
fig=plt.gcf()
fig.set_size_inches(12,12)
plt.show()
```

10.Проверить несколько гипотез для данных, чтобы понять почему сотрудники покидают компанию

```python
#Нулевая гипотеза о том, что у уволенных и работающих нет разницы в количестве проектов, проводимых за время работы
data['projectsByTime'] = data['average_montly_hours']/data['number_project']
fired = data[data['left']==1]['projectsByTime']
working = data[data['left']==0]['projectsByTime']

fired = fired.sample(500)
working = working.sample(500)
```

```python
statistics.pvariance(fired), statistics.pvariance(working)
```

```python
stats.ttest_ind(fired, working, equal_var=False)
```

Нулевая гипотеза отвергается, значит, одна из причин увольнения - трата большого кол-ва времени на малое кол-во проектов, т.е. нет разнообразия или развития

```python
#Нулевая гипотеза о том, что удовлетворенность работой не связана с увольнениями
fired = data[data['left']==1]['satisfaction_level']
working = data[data['left']==0]['satisfaction_level']

fired = fired.sample(500)
working = working.sample(500)
```

```python
statistics.pvariance(fired), statistics.pvariance(working)
```

```python
stats.ttest_ind(fired, working, equal_var=False)
```

Нулевая гипотеза отвергается, значит, одна из причин увольнения - недовольство работой

```python
#Нулевая гипотеза о том, что соотношение зп и времени работы не причина увольнения
data['salaryBywork'] = data['salaryInt']/data['average_montly_hours']
fired = data[data['left']==1]['salaryBywork']
working = data[data['left']==0]['salaryBywork']

fired = fired.sample(500)
working = working.sample(500)
```

```python
statistics.pvariance(fired), statistics.pvariance(working)
```

```python
stats.ttest_ind(fired, working, equal_var=False)
```

Нулевая гипотеза отвергается, значит, одна из причин увольнения - недовольство соотношением времени работы и зарплатой
